### Defining axes, ticks, and grids

```python
x=range(1,10)
y=[1,2,3,4,0,4,3,2,1]
fig=plt.figure()
ax= fig.add_axes([.1,.1,1,1])
ax.plot(x,y)
```

```python
x=range(1,10)
y=[1,2,3,4,0,4,3,2,1]
fig=plt.figure()
ax= fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])
#limits 7:45 in video



ax.plot(x,y)
```
